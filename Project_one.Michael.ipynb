```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('Resources/Poverty_Data_2018.csv', encoding="unicode_escape")
df
```

```python
df=df.rename(columns={'Unnamed: 6':'Poverty Percent, All Ages'})
```

```python
df=df.loc[4:]
df
```

```python
df=df.rename(columns={'Unnamed: 2':'State'})
df
```

```python
df=df[['State','Poverty Percent, All Ages']]
df
```

```python
df=df.reset_index(drop=True)
df
```

```python
plt.hist(df['Poverty Percent, All Ages'].astype(float),bins=10,rwidth=.5)
plt.title('Poverty Rate Histogram')
plt.xlabel('Poverty Rate Percentage')
plt.ylabel('State Count')
```

```python

df['Poverty Percent, All Ages']=df['Poverty Percent, All Ages'].astype(float)
temp=df.sort_values('Poverty Percent, All Ages')
```

```python
fig,ax=plt.subplots(figsize=(10,3))
ax=plt.bar(temp.State,temp['Poverty Percent, All Ages'])
plt.xticks(rotation=90,fontsize=10)
```
