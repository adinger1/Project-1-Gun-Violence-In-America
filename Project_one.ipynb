```python
# import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Open first CSV
state_count_df = pd.read_csv("Resources/state_count_all_incidents.csv")
state_count_df.head()
```

```python
# Open second CSV
education_df = pd.read_csv("Resources/David_Education_Pop.csv")
education_df.head()
```

```python
 # Rename column in Second CSV to State for merging
education_df = education_df.rename(columns={'Name': 'State'})
education_df.head()
```

```python
# Merge DataFrames
merged_data_df = pd.merge(state_count_df, education_df, on='State')
merged_data_df.head()
```

```python
# Change column in David's DataFrame to numeric value
merged_data_df["Less than high school graduate"]=merged_data_df.iloc[:,[3]].replace('\%','',regex=True).astype(float)
New_Series=merged_data_df["Less than high school graduate"]
New_Series
```

```python
# Create 4 DataFrame Series for plotting 4 graphs in one line
df1 = merged_data_df.iloc[:,[6]]
df1['# incidents'] = merged_data_df['# Incidents']

df2 = merged_data_df.iloc[:,[8]]
df2['# incidents'] = merged_data_df['# Incidents']

df3 = merged_data_df.iloc[:,[9]]
df3['# incidents'] = merged_data_df['# Incidents']

df4 = merged_data_df.iloc[:,[10]]
df4['# incidents'] = merged_data_df['# Incidents']



```

```python
# Plot
fig, axs = plt.subplots(nrows=2, ncols=2)
df1.plot(ax=axs[0,0])
df2.plot(ax=axs[0,1])
df3.plot(ax=axs[1,0])
df4.plot(ax=axs[1,1])
```
